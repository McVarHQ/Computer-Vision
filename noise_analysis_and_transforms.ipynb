#Task 1: Determine the type of noise of the given input image and suggest a suitable method for reducing the noise in the image.

#Task 2: For any of the standard test image perform the following geometric transformations: Rotate the image by an angle of 45 deg clockwise

```python
img = cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.title('Original(Grayscale)')
img.shape
```

```python
rot_matrix = np.float32([
    [np.cos(np.pi/4), -np.sin(np.pi/4), 0],
    [np.sin(np.pi/4), np.cos(np.pi/4), 0]
])

transformed_image = cv2.warpAffine(img, rot_matrix, (img.shape[1], img.shape[0]))
```

```python
plt.imshow(transformed_image, cmap='gray')
plt.title('Rotated')
img.shape
```

Rotate Around Center

```python
height, width = img.shape[:2]
center = (width/2, height/2)

rot_matrix = cv2.getRotationMatrix2D(center=center, angle=-45, scale=1)
transformed_image = cv2.warpAffine(img, rot_matrix, (img.shape[1], img.shape[0]))
```

```python
plt.imshow(transformed_image, cmap='gray')
plt.title('Rotated')
img.shape
```

###Inference:
Using the Affine tranform matrix turns the image about the origin.
Using the getrotationalmatrix function also adds a tranformational value to the matrix to offset it so that the object rotates about its center only

#Task 3: Rotate and scale the image by an angle of -90 deg and scale it to factor of 2 respectively

```python
height, width = img.shape[:2]
center = (width/2, height/2)

rot_matrix = cv2.getRotationMatrix2D(center=center, angle=-90, scale=2)
transformed_image = cv2.warpAffine(img, rot_matrix, (img.shape[1], img.shape[0]))
```

```python
plt.imshow(transformed_image, cmap='gray')
plt.title('Rotated and Scaled')
img.shape
```

###Inference:
This rotates and scales the image. The scaling accounts for the empty space created while rotating
