# Task 1: Perform image smoothening of an input image based on ideal filters(take kernels of different size ) and gaussian filters.


Ideal Filters

```python
kernel_size_1 = 3
kernel_1 = np.ones((kernel_size_1, kernel_size_1), np.float32) / (kernel_size_1 * kernel_size_1)

kernel_size_2 = 25
kernel_2 = np.ones((kernel_size_2, kernel_size_2), np.float32) / (kernel_size_2 * kernel_size_2)

# Apply convolution with the ideal filter kernels
smooth_image_1 = cv2.filter2D(image, -1, kernel_1)
smooth_image_2 = cv2.filter2D(image, -1, kernel_2)

# Display the smoothened images
plt.subplot(2,2,1)
plt.imshow(image,"gray")
plt.title("Orignal Image")
plt.subplot(2,2,2)
plt.imshow(smooth_image_1,"gray")
plt.title("Kernel Size = 3")
plt.subplot(2,2,3)
plt.imshow(smooth_image_2,"gray")
plt.title("Kernel Size = 25")
plt.tight_layout()
plt.show()
```

Gaussian Filters

```python
# Apply Gaussian blur with different kernel sizes

image = cv2.imread("baboon.png", 0)

smooth_image_gaussian_1 = cv2.GaussianBlur(image, (13, 13), 0)
smooth_image_gaussian_2 = cv2.GaussianBlur(image, (17, 17), 0)

plt.subplot(2,2,1)
plt.imshow(image,"gray")
plt.title("Orignal")
plt.subplot(2,2,2)
plt.imshow(smooth_image_gaussian_1,"gray")
plt.title("Kernel = 13")
plt.subplot(2,2,3)
plt.imshow(smooth_image_gaussian_2,"gray")
plt.title("kernel = 17")
plt.tight_layout()
plt.show()
```

###Inference:
The more the kernel size the more blurry the image gets as edges lose more and more definiton

# Task 3: Perform image sharpening of an input image based on ideal filters (take kernel values ).

#Task 3: Perform spatial filtering based on Order statistics filter (Median, Min, Max filters)

```python
# Read image
image = cv2.imread('baboon.png', 0)

# Median filter
median = cv2.medianBlur(image, 5)

# Min filter
kernel = np.ones((5,5),np.uint8)
min = cv2.erode(image,kernel)

# Max filter
max = cv2.dilate(image,kernel)
```

```python
plt.subplot(2,2,1)
plt.imshow(image,"gray")
plt.title("Orignal")

plt.subplot(2,2,2)
plt.imshow(median,"gray")
plt.title("Median Filter")

plt.subplot(2,2,3)
plt.imshow(min,"gray")
plt.title("Min Filter")

plt.subplot(2,2,4)
plt.imshow(max,"gray")
plt.title("Max Filter")

plt.tight_layout()
plt.show()
```

###Inference:
Median filter replaces each pixel with the median value of the neighboring pixels. It's effective for reducing noise and preserving edges. The median filter is particularly good for salt and pepper noise.

Minimum filter eplaces the central pixel with the darkest pixel in the neighborhood.

Maximum filter replaces the central pixel with the lightest pixel in the neighborhood. It's also known as the dilation filter.
