# Task 1: Perform Contrast enhancement using Histogram equalization


```python
image.shape
```

```python
# Calculate histogram
hist, bins = np.histogram(image.flatten(), 256, [0, 256])

# Plot histogram
plt.figure(figsize=(8, 4))
plt.hist(image.flatten(), 256, [0, 256], color='red')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Histogram of Original Image')
plt.show()
```

```python
equalized_image = cv2.equalizeHist(image)
```

```python
# Calculate histogram of equalized image
equalized_hist, _ = np.histogram(equalized_image.flatten(), 256, [0, 256])

# Plot histogram of equalized image
plt.figure(figsize=(8, 4))
plt.hist(equalized_image.flatten(), 256, [0, 256], color='red')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Histogram of Equalized Image')
plt.show()
```

```python
plt.imshow(equalized_image,"gray")
plt.title("Equalized")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.imshow(image,"gray")
plt.title("Orignal")
plt.subplot(2,2,2)
plt.title("Equalized")
plt.imshow(equalized_image,"gray")
plt.show()
```

Inference :

> Histogram Equalization is a computer image processing technique
> used to improve contrast in images. This method effectively
> spreaded out the most frequent intensity values,
> i.e. stretching out the intensity range of the image.
> we get a enhanced image as shown abhove


# Task 2: Perform image smoothening of an input image based on ideal filters(take kernels of different size ) and gaussian filters.


## 2.1 Apply ideal filters


```python
image = cv2.imread('image1.jpg', 0)


kernel_size_1 = 3
kernel_1 = np.ones((kernel_size_1, kernel_size_1), np.float32) / (kernel_size_1 * kernel_size_1)

kernel_size_2 = 25
kernel_2 = np.ones((kernel_size_2, kernel_size_2), np.float32) / (kernel_size_2 * kernel_size_2)

# Apply convolution with the ideal filter kernels
smooth_image_1 = cv2.filter2D(image, -1, kernel_1)
smooth_image_2 = cv2.filter2D(image, -1, kernel_2)

# Display the smoothened images
plt.subplot(2,2,1)
plt.imshow(image,"gray")
plt.title("Orignal Image")
plt.subplot(2,2,2)
plt.imshow(smooth_image_1,"gray")
plt.title("Kernel Size = 3")
plt.subplot(2,2,3)
plt.imshow(smooth_image_2,"gray")
plt.title("Kernel Size = 25")
plt.show()
```

## 2.3 Apply Gaussian filters


```python
# Apply Gaussian blur with different kernel sizes

image = cv2.imread("image1.jpg", 0)

smooth_image_gaussian_1 = cv2.GaussianBlur(image, (13, 13), 0)
smooth_image_gaussian_2 = cv2.GaussianBlur(image, (17, 17), 0)

plt.subplot(2,2,1)
plt.imshow(image,"gray")
plt.title("Orignal")
plt.subplot(2,2,2)
plt.imshow(smooth_image_gaussian_1,"gray")
plt.title("Kernel = 13")
plt.subplot(2,2,3)
plt.imshow(smooth_image_gaussian_2,"gray")
plt.title("kernel = 17")
plt.show()
```

# Task 3: Perform image sharpening of an input image based on ideal filters (take kernel values ).


```python
image = cv2.imread('image1.jpg',0)

# deal filter kernel for image sharpening
kernel_sharpening = np.array([
    [0,-1,0],
    [-1,4,-1],
    [0,-1,0]
])


sharpened_image = cv2.filter2D(image, -1, kernel_sharpening)

plt.subplot(2,2,1)
plt.imshow(image,"gray")
plt.title("Orignal")
plt.subplot(2,2,2)
plt.imshow(sharpened_image,"gray")
plt.title("Kernel1")
plt.show()
```
