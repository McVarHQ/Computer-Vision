**_Image Segmentation_**


# Setup


```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image1 = cv2.imread("image1.jpg", 0)
```

```python
plt.imshow(image1, "gray")
plt.title("Original")
```

# Image Segmentation using Simple Thresholding


```python
threshold = 127  # Adjust threshold value as needed
simple_threshold_img = np.where(image1 < threshold, 0, 255)
plt.imshow(simple_threshold_img, "gray")
plt.title("Simple Thresholding")
```

# Image Segmentation using Histogram-based Thresholding


```python
# Calculate histogram
hist_image1 = cv2.equalizeHist(image1)
hist, bins = np.histogram(hist_image1.flatten(), 256, [0, 256])

# Plot histogram
plt.figure(figsize=(8, 4))
plt.hist(hist_image1.flatten(), 256, [0, 256])
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.title("Histogram of Original Image")
plt.show()
```

```python
thresh, dst = cv2.threshold(image1, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(dst, "gray")
plt.title("Global Thresholding")
```

# Conclusion


```python
plt.subplot(2, 2, 1)
plt.imshow(image1, "gray")
plt.title("Original")
plt.subplot(2, 2, 2)
plt.imshow(simple_threshold_img, "gray")
plt.title("Simple Thresholding")
plt.subplot(2, 2, 3)
plt.imshow(dst, "gray")
plt.title("Global Thresholding")
```
