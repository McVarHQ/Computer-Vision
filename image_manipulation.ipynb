```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
cv.__version__
```

```python
# import the image
img1bgr = cv.imread("airplane.png")
```

```python
# Get the attributes of the image
type(img1bgr)
```

```python
img1bgr.shape
```

```python
# Plot the image
plt.imshow(img1bgr)
plt.title("BGR Image")
```

```python
# Correct the Color Space
img1rgb = cv.cvtColor(img1bgr, cv.COLOR_BGR2RGB)
```

```python
# Plot the rgb image
plt.subplot(1,2,1)
plt.imshow(img1bgr)
plt.title("BGR Airplane")
plt.subplot(1,2,2)
plt.imshow(img1rgb)
plt.title("RGB Airplane")
```

```python
r_plane = img1rgb[:, :, 0]
g_plane = img1rgb[:, :, 1]
b_plane = img1rgb[:, :, 2]
```

```python
# Color Space 1
r_plane.min(), r_plane.max(), r_plane.mean()
```

```python
# Color Space 2
g_plane.min(), g_plane.max(), g_plane.mean()
```

```python
# Color Space 3
b_plane.min(), b_plane.max(), b_plane.mean()
```

```python
# Plot the Cropped image
cropped_img1 = img1rgb[100:300, 20:480]
plt.imshow(cropped_img1)
plt.title("Cropped Image")
```

```python
# Convert to Graymap
img1gray = cv.cvtColor(img1bgr, cv.COLOR_BGR2GRAY)

plt.imshow(img1gray, cmap='gray')
plt.title("Grayscale Image")
```

```python
img1gray.shape
```

```python
K = 30

img1brightRGB = img1rgb + K

# Convert to HSV
img1hsv = cv.cvtColor(img1bgr, cv.COLOR_BGR2HSV)
img1brightHSV = img1hsv
value_channel = img1hsv[:, :, 2]
value_brighter = value_channel + K
value_brighter = np.clip(value_brighter, 0, 255)
img1brightHSV[:, :, 2] = value_brighter
img1brightHSV = cv.cvtColor(img1brightHSV, cv.COLOR_HSV2RGB)

plt.subplot(1,3,1)
plt.imshow(img1brightRGB)
plt.title("Brighter w/o HSV")
plt.subplot(1,3,2)
plt.imshow(img1rgb)
plt.title("RGB Airplane")
plt.subplot(1,3,3)
plt.imshow(img1brightHSV)
plt.title("Brighter with HSV")
```
