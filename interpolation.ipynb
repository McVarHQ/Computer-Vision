```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

#Task 1: Take any sample image of size 256x256 and up sample it to size of 512x512 based on a) Nearest Neighbor b) Linear, c) Bilinear interpolation techniques. Comment on the nature of the upsampled images obtained by these methods.

```python
img = cv2.imread('lena_small.png', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img, cmap='gray')
plt.title('Original(Grayscale)')
```

```python
img.shape
```

Apply different Interpolation

```python
resized_nearest = cv2.resize(img, (512, 512), interpolation=cv2.INTER_NEAREST)

resized_linear = cv2.resize(img, (512, 512), interpolation=cv2.INTER_LINEAR)

resized_bilinear = cv2.resize(img, (512, 512), interpolation=cv2.INTER_AREA)
```

```python
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original(Grayscale)')

plt.subplot(2, 2, 2)
plt.imshow(resized_nearest, cmap='gray')
plt.title('Nearest Neighbour')

plt.subplot(2, 2, 3)
plt.imshow(resized_linear, cmap='gray')
plt.title('Linear')

plt.subplot(2, 2, 4)
plt.imshow(resized_bilinear, cmap='gray')
plt.title('Bilinear')

plt.tight_layout()
plt.show()
```

###Inference:
Nearest neighbor interpolation is the fastest and most efficient, but it produces the lowest quality images. It's ideal for real-time applications where speed is critical. Nearest neighbor interpolation selects the value of the nearest point, and it retains the exact values of the original pixels.

Linear interpolation is simple, but it may not be accurate enough for complex data. Linear interpolation uses a linear polynomial to construct new data points based on the two nearest points.

Bilinear interpolation offers a balance between quality and performance. It considers the closest 2x2 neighborhood of known pixel values surrounding the unknown pixel, and then takes a weighted average of these 4 pixels to arrive at its final interpolated value. Bilinear interpolation produces smoother looking images than nearest neighbor interpolation.
